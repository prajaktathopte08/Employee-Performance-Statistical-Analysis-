```python
# Mean (Average)
# I calculated the average monthly salary to understand the general salary level of employees.
mean_salary = np.mean(data["Monthly_Salary"])
print("Mean Monthly Salary:", mean_salary)
```

```python
#  Median (Middle Value)
# I calculated the median salary to find the middle salary value.
median_salary = np.median(data["Monthly_Salary"])
print(median_salary)
```

```python
# Mode (Most Frequent Value)
# I found the most frequently occurring salary to understand common pay ranges.
values, counts = np.unique(data["Monthly_Salary"], return_counts=True)
mode_salary = values[np.argmax(counts)]
print(mode_salary)
```

```python
# Variance helped me understand how much salaries differ from the average.
# Your variance value is very large, which means:
# Employee salaries vary a lot — some earn much less and some earn much more.
salary_variance = round(np.var(data["Monthly_Salary"]),2)
print(salary_variance)
```

```python
# It tells us on average, how much salaries differ from the mean salary
salary_std = round(np.std(data["Monthly_Salary"]),2)
print(salary_std)
```

```python
# Percentiles & Quartiles
q1 = np.percentile(data["Monthly_Salary"], 25)  #Q1 → Bottom 25% salaries
q2 = np.percentile(data["Monthly_Salary"], 50)  #Q2 → Median
q3 = np.percentile(data["Monthly_Salary"], 75)  # Q3 → Top 25% salaries
```

```python
# IQR: Spread of middle 50% data
# Range showed the difference between the highest and lowest salary
# IQR showed the spread of the middle 50% of salaries
# IQR is useful because it ignores extreme salary values.

salary_range = np.max(data["Monthly_Salary"]) - np.min(data["Monthly_Salary"])
iqr = q3 - q1
```

```python
# Correlation Analysis
# It show the negative cor-reletion becouse the experiance increases but not salary 
correlation = round(np.corrcoef(
    data['Experience_Years'],   # Experience
    data['Monthly_Salary']    # Salary
)[0, 1],2)

print(correlation)
```

```python
# Find the Minimum and Maximum Monthly Salary
min_salary = np.min(data["Monthly_Salary"])
max_salary = np.max(data["Monthly_Salary"])

print("Minimum Salary:", min_salary)
print("Maximum Salary:", max_salary)
```

```python
# Calculate Skewness of Monthly Salary
# “The skewness value is close to zero, which means the salary distribution is almost symmetrical
#  and not heavily biased toward either low or high salaries.”
mean = np.mean(data["Monthly_Salary"])
median = np.median(data["Monthly_Salary"])
std = np.std(data["Monthly_Salary"])

skewness = round((mean - median) / std,4)
print("Skewness:", skewness)
```

```python
# Find Employees Earning Above Average Salary
# How many employees earn more than the average salary
above_avg = data[data["Monthly_Salary"]> np.mean(data["Monthly_Salary"])]
print("Employees above average salary:", len(above_avg))
```
